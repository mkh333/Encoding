# Ordinal Encoder

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
cities = df8.region.unique()
```

```python
cities
```

```python
ordinal = OrdinalEncoder(categories=[cities])
```

```python
ordinal
```

```python
encoded = ordinal.fit_transform(df8[['region']])
```

```python
encoded
```

```python
pd.DataFrame(encoded, columns=['region'])
```

```python
df8['region'] = pd.DataFrame(encoded, columns=['region'])
```

```python
df8.head()
```

```python
for col in df8.columns:
    if is_numeric_dtype(df8[col]):
        continue
    unq = df8[col].unique()
    ordinal = OrdinalEncoder(categories=[unq])
    encoded = ordinal.fit_transform(df8[[col]])
    df8[col] = pd.DataFrame(encoded, columns=[col])
    
    
```
